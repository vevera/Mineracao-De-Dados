## Convertendo DateTime em Data e Hora convencional

```python
lei_rouanet_csv = pd.read_csv('dados_lei_rouanet_08_12.csv')
```

```python
from datetime import datetime
```

```python
horas = []
dias = []
```

```python
for row in lei_rouanet_csv.itertuples():
    data_time = row.data_time.replace(".000Z","")
    date_time_obj = datetime.strptime(data_time, '%Y-%m-%dT%H:%M:%S')
    dias.append(date_time_obj.strftime("%d/%m/%Y"))
    horas.append(date_time_obj.strftime("%H:%M:%S"))
    
print("Done!")
```

```python
len(horas)
```

```python
len(dias)
```

```python
lei_rouanet_csv["dia"] = dias
```

```python
lei_rouanet_csv["hora"] = horas
```

```python
lei_rouanet_csv = lei_rouanet_csv.drop(columns=["data_time"])
```

```python
lei_rouanet_csv.to_csv('lei_rouanet_final.csv', index=False)
```
